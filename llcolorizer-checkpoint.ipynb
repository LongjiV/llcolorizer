Colorizer function

```python
# define add_color function
def add_color(image):
    
    #convert image from BGR to RGB
    rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    #split values into variables
    r, g, b = cv2.split(rgb_image)

    # multiply through all values
    np.multiply(r, 1.5, out=r, casting="unsafe")
    np.multiply(g, .75, out=g, casting="unsafe")
    np.multiply(b, 1.25, out=b, casting="unsafe")

    #merge values together
    after = cv2.merge([r, g, b])

    #plot image
    fig = plt.figure()
    ax1 = fig.add_subplot(1, 2, 1)
    ax2 = fig.add_subplot(1, 2, 2)
    
    fig1 = ax1.imshow(image)
    fig2 = ax2.imshow(after)
    
    return fig1, fig2
```

```python
add_color(new_image)
```

```python
def skin_detector(image):
    
    #convert image from BGR to HSV
    new_image2 = cv2.imread(r"C:\Users\CCECC\Documents\GitHub\Cohort4\intermediate\computer-vision\week03\notebook\teju.jpg")    
    hsv_image = cv2.cvtColor(new_image2, cv2.COLOR_BGR2HSV)
    
    #define upper and lower limit for skin pixels
    lower = np.array([0,48,80],np.uint8)
    upper = np.array([20,255,255],np.uint8)
    
    skin_region = cv2.inRange(hsv_image, lower, upper)
    
    skin = cv2.bitwise_and(new_image2, new_image2, mask = skin_region)

    return plt.imshow(skin)
```
